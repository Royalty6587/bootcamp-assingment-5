```python
# Import libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import warnings

# Suppress FutureWarnings
warnings.simplefilter(action='ignore', category=FutureWarning)

# Task 1: Load the penguins dataset and perform data preprocessing
dataset = sns.load_dataset("penguins")

# Basic data preprocessing
dataset.fillna(dataset.mean(numeric_only=True), inplace=True)
dataset.dropna(inplace=True)
dataset.reset_index(drop=True, inplace=True)

# Display the first few rows to check the dataset
print(dataset.head())

# Task 2: Plotting relationships between different columns using Matplotlib
plt.figure(figsize=(10, 6))
plt.scatter(dataset['flipper_length_mm'], dataset['body_mass_g'], c='blue', label="Flipper Length vs Body Mass")
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Body Mass (g)')
plt.title('Relationship between Flipper Length and Body Mass')
plt.legend()
plt.show()

# Task 3: Plotting pair plot using Seaborn
sns.pairplot(dataset, hue='species', palette="viridis")
plt.show()

sns.pairplot(dataset, hue='sex', palette="coolwarm")
plt.show()
```

```python

import pandas as pd

# Load the dataset/
df = pd.read_csv('checkin_checkout_history_updated.csv')

# Display the first few rows
print(df.head())
# Check the summary and structure of the dataset
print(df.info())
# Check for basic statistics
print(df.describe())
```

```python
# Check for missing values
print(df.isnull().sum())

# Example: Fill missing numerical values with the mean
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
# Check for duplicates
print(df.duplicated().sum())

# Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot histograms for numerical features
df.hist(bins=30, figsize=(15, 10))
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

```
